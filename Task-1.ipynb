```python
#Task 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load World Bank dataset
df = pd.read_csv(r"C:\Users\jg275\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_127006\API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv", skiprows=4)
india_df = df[df['Country Name'] == 'India']
population_2022 = india_df['2022'].values[0]

age_groups = ['0–20 Years', '21–64 Years', '65+ Years']
percentages = [36.1, 57.0, 6.9]
pop_millions = [(population_2022 * p / 100) / 1e6 for p in percentages]


colors = ['#4E79A7', '#A0CBE8', '#B2B2B2']  # navy blue, soft blue, grey

plt.figure(figsize=(10, 6))
bars = plt.barh(age_groups, pop_millions, color=colors)

# Add text inside bars
for bar, pop, pct in zip(bars, pop_millions, percentages):
    plt.text(bar.get_width() - 30, bar.get_y() + bar.get_height()/2,
             f'{pop:.0f} Mn\n{pct}%', va='center', ha='right',
             color='white' if bar.get_width() > 100 else 'black',
             fontsize=12, fontweight='bold')

# Styling
plt.title("India's Population Distribution by Age Group (2022)", fontsize=15, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=12)
plt.xlim(0, max(pop_millions) + 100)
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
india_df = df[df['Country Name'] == 'India']
population_2022 = india_df['2022'].values[0]

age_groups = [
    (0, 20, 36.1),
    (21, 64, 57.0),
    (65, 100, 6.9)
]

total_sample = 100000
ages = []

for min_age, max_age, percent in age_groups:
    count = int((percent / 100) * total_sample)
    sampled = np.random.randint(min_age, max_age + 1, count)
    ages.extend(sampled)

# 
histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=range(0, 105, 5), color="#4E79A7", edgecolor='white')
plt.title("India Age Distribution (Simulated Histogram, 2022)", fontsize=15)
plt.xlabel("Age")
plt.ylabel("People (Sample of 100K)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
